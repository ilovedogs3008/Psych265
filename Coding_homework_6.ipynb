Task 8: I estimate that the dominant oscillation for ACC LFP is roughly around 3.5 Hz, while the dominant oscillation for the LC LFP would be around 4 Hz. While the oscillation of these two types of cells are relatively similar, the characteristics of the waves do differ drastically. One interesting observation about the relationship between ACC LFPs and LC LFPs is that they apper to have alternating amplitudes; when the amplitude of ACC is high, the amplitude of LC tends to be low.

```python
#Task 9: Calculate the coherance for acc_hits and lc_hits

dt = time[1] - time[0]   
K = acc_hits.shape[0]              
N = acc_hits.shape[1]               
ac = np.zeros([2 * N - 1])  

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0) 
Syy = scale * (yf * yf.conj()).mean(0) 
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                   
plt.plot(f, cohr.real)                   
plt.xlim([0,  100])                    
plt.ylim([0,1.5])                    
plt.xlabel('Frequency [Hz]')               
plt.ylabel('Coherence')
plt.title('Coherence between ACC_hits and LC_hits')
```

```python
#Task 9 continued

x_fa = acc_fa[1,:] - np.mean(acc_fa[1,:])
y_fa = lc_fa[1,:] - np.mean(lc_fa[1,:])
xf_fa = np.array([np.fft.rfft(x_fa-x_fa.mean()) for x_fa in acc_fa])
yf_fa = np.array([np.fft.rfft(y_fa-y_fa.mean()) for y_fa in lc_fa])

T = time[-1]
dt = time[1] - time[0]             
scale = 2 * dt**2 / T
Sxx = scale * (xf_fa * xf_fa.conj()).mean(0)  
Syy = scale * (yf_fa * yf_fa.conj()).mean(0) 
Sxy = scale * (xf_fa * yf_fa.conj()).mean(0) 


cohr2 = np.abs(Sxy)/(np.sqrt(Sxx)* np.sqrt(Syy))
N2 = acc_fa.shape[1]


f_fa = np.fft.rfftfreq(N2, dt)
plt.plot(f_fa, cohr2.real)
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC')
```
