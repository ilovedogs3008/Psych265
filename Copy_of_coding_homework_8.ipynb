```python
#Importing the modules for the GLM
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
#Task 10: making a new variable with shock onset and tone onset
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0
```

```python
#Task 10 continued: Run a for loop to create a GLM for the conditioned session
r_model_conditioning = np.zeros(62)
predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array

for i in range(62):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_conditioning = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r_index_conditioning.statistic
```

```python
#Task 11: do the previous steps but for the extinction_traces
shock_array = np.zeros(14178)
for i in range(14178):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14178)
for i in range(14178):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

extinction_predictors = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
extinction_predictors['shock onset'] = shock_array
```

```python
#Task 11 continued: run a for loop to create the GLM for the extinction session
r_model_extinction = np.zeros(62)

for i in range(62):
  trace_temp = extinction_traces[:,i]
  model3 = sm.GLM(trace_temp, extinction_predictors, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_extinction = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r_index_extinction.statistic
     
```

```python
#Task 12: Correlate r_index_extinction and r_index_conditioning using pearsonr function
#I ran it with the index variables at first and that resulted in a statistic of 1 and a p-value of 1. That seemed wrong so I
#Performed the same person correlation using the model variables

pearsonr(r_model_extinction, r_model_conditioning)
```

```python
#Task 13: make a scatter plot.
plt.scatter(r_model_conditioning, r_model_extinction)
plt.xlabel("Coefficients for the Conditioning Session")
plt.ylabel("Coefficients for the Extinction Session")
     
```
