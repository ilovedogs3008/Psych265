```python
#Task 2: Import the relevant modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Task 3: Importing the datasets.

url1 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/spikes_left_trial1.csv'
spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/spikes_right_trial1.csv'
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')

url3 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/spikes_left_trial2.csv'
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')

url4 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/spikes_right_trial2.csv'
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
#Task 4: print the lenght of each array

print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
#Task 5: Firing rate of neuron at different trials
print((len(spikes_right_trial1))/30)
print((len(spikes_right_trial2))/30)
print((len(spikes_left_trial1))/30)
print((len(spikes_left_trial2))/30)
```

Task 6: This neuron appears to fire more during the right trial. However, the firing rates are relatively similar. 

```python
#Task 7: 
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.')
ax[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
ax[0].set_xlim([0,30])
ax[0].set_ylim([0, 3])
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Spike per trial type')
ax[0].legend({'Right', 'Left'})
ax[0].set_title('Spike Train for Trial 1')
ax[0].set_yticks([])

ax[1].plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.') 
ax[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
ax[1].set_xlim([0,30])
ax[1].set_ylim([0, 3])
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Spike per trial type')
ax[1].legend({'Right', 'Left'})
ax[1].set_title('Spike Train for Trial 2')
ax[1].set_yticks([])
```

Task 8: This neuron appears more bursty, with it having periods where it is rather inactive and periods where activation is quick in sucession. Considering the data from both trial 1 and trial 2, it appears that the neuron is more active during right turn trials. 

```python
#Task 9: Calculate the ISI for each of the four spike arrays.
Right1ISI = np.diff(spikes_right_trial1)
Left1ISI = np.diff(spikes_left_trial1)
Right2ISI = np.diff(spikes_right_trial2)
Left2ISI = np.diff(spikes_left_trial2)
```

```python
#Task 10: Make a Histogram of the ISIs
bins = np.arange(0, 2, 0.01)
fig, ax = plt.subplots(2,2, figsize=(15,10))
ax[0,0].hist(Right1ISI, bins)
ax[0,0].set_xlabel('Time [s]') 
ax[0,0].set_ylabel('ISI')
ax[0,0].set_title("ISI Values for First Right Turn Trial")

ax[0,1].hist(Right2ISI, bins)
ax[0,1].set_xlabel('Time [s]') 
ax[0,1].set_ylabel('ISI')
ax[0,1].set_title("ISI Values for Second Right Turn Trial")

ax[1,0].hist(Left1ISI, bins)
ax[1,0].set_xlabel('Time [s]')
ax[1,0].set_ylabel('ISI')
ax[1,0].set_title("ISI Values for First Left Turn Trial")

ax[1,1].hist(Left2ISI, bins)
ax[1,1].set_xlabel('Time [s]') 
ax[1,1].set_ylabel('ISI')
ax[1,1].set_title("ISI Values for Second Left Turn Trial")
plt.show()
```

Task 11: As evident by the histograms, the neuron appears to fire in bursts, with groupings going as high as 7 individual action potentials. This "burstiness" is especially evident in the right turn trials.

```python
#Task 12: Calculate the firing rate for each of the four arrays

binsize = 1
firing_bins = np.arange(0,30,binsize)                                        
firing_right1 = np.zeros(len(firing_bins)-1)                                                           
for i in range(len(firing_bins)-1):                                            
  spk_index = (spikes_right_trial1 > firing_bins[i]) & (spikes_right_trial1 < firing_bins[i+1])          
  true_values = spikes_right_trial1[spk_index]                                         
  firing_right1[i] = (len(true_values))/binsize

firing_bins = np.arange(0,30,binsize)                                        
firing_right2 = np.zeros(len(firing_bins)-1)                                                           
for i in range(len(firing_bins)-1):                                            
  spk_index = (spikes_right_trial2 > firing_bins[i]) & (spikes_right_trial2 < firing_bins[i+1])          
  true_values = spikes_right_trial2[spk_index]                                         
  firing_right2[i] = (len(true_values))/binsize

firing_bins = np.arange(0,30,binsize)                                        
firing_left1 = np.zeros(len(firing_bins)-1)                                                           
for i in range(len(firing_bins)-1):                                            
  spk_index = (spikes_left_trial1 > firing_bins[i]) & (spikes_left_trial1 < firing_bins[i+1])          
  true_values = spikes_left_trial1[spk_index]                                         
  firing_left1[i] = (len(true_values))/binsize

firing_bins = np.arange(0,30,binsize)                                        
firing_left2 = np.zeros(len(firing_bins)-1)                                                           
for i in range(len(firing_bins)-1):                                            
  spk_index = (spikes_left_trial2 > firing_bins[i]) & (spikes_left_trial2 < firing_bins[i+1])          
  true_values = spikes_left_trial2[spk_index]                                         
  firing_left2[i] = (len(true_values))/binsize
```

```python
#Task 12 Continued: Create four subplots

fig, ax = plt.subplots(2,2,figsize=(10,10))
ax[0,0].plot(firing_bins[:-1], firing_right1, "b")
ax[0,0].set_xlabel('Time (s)')
ax[0,0].set_ylabel('Firing Rate (Hz)')
ax[0,0].set_title("Firing Rate for First Right Turn Trial")

ax[0,1].plot(firing_bins[:-1], firing_right2, "b")
ax[0,1].set_xlabel('Time (s)')
ax[0,1].set_ylabel('Firing Rate (Hz)')
ax[0,1].set_title("Firing Rate for Second Right Turn Trial")

ax[1,0].plot(firing_bins[:-1], firing_left1, "r")
ax[1,0].set_xlabel('Time (s)')
ax[1,0].set_ylabel('Firing Rate (Hz)')
ax[1,0].set_title("Firing Rate for First Left Turn Trial")

ax[1,1].plot(firing_bins[:-1], firing_left2, "r")
ax[1,1].set_xlabel('Time (s)')
ax[1,1].set_ylabel('Firing Rate (Hz)')
ax[1,1].set_title("Firing Rate for Second Left Turn Trial")


```

Task 13: the firing rate for the neuron appers to differ between the left turn and right turn trials. As seen in the graphs, the right turn firing trials tend to fire more during the first half of the 20 second waiting period while the left turn trials appear to fire more during the latter half of the 30-second waiting period. However, this pattern is more evident during the second trial rather than the first, where the distribution of the left turn trial is a bit more normally distributed.
