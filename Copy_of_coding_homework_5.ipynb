```python
#Task 2: import the relevant modules

import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Task 3: Importing the relevant files

link1 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(link1, delimiter = ',')

link2 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(link2, delimiter = ',')
```

```python
#Task 4: print the shape of each array

print(acc_hits.shape, acc_fa.shape)
```

```python
#Task 5: creating a time variable

time = np.linspace(0,4,8001)
```

```python
#Task 6: Create an expression to determine the sampling rate of the data.
print(8001/time[-1])
```

```python
#Task 7: Plotting the first trials of acc_hits and acc_fa
fig, trials = plt.subplots(2,1,figsize=(10,12))

trials[0].plot(time, acc_hits[0])
trials[0].set_xlabel('Time')
trials[0].set_ylabel('Amplitude')
trials[0].set_title('Hits')


trials[1].plot(time,acc_fa[0])
trials[1].set_xlabel('Time')
trials[1].set_ylabel('Amplitude')
trials[1].set_title('False Alarm')
```
