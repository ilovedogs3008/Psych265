```python
#Task 2: Import numpy, matplot, and scipy; print done to ensure the program ran successfully 

import numpy as np

import matplotlib.pyplot as plt

import scipy.stats as stats

print('Done')
```

```python
#Task 3: import v_rest1 and v_rest2 as numpy arrays; set them to the variables  v_rest1 and v_rest2, respectively; print the data to double check

url = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/vrest_1'
v_rest1 = np.loadtxt(url)

url2 = 'https://raw.githubusercontent.com/ilovedogs3008/Psych265/main/vrest_2%20-%20vrest_2.csv'
v_rest2 = np.loadtxt(url2)

print('done')

v_rest1, v_rest2
```

```python
#Task 4: a) Find the means of v_rest1 and 2; b) perform a t-test; c) histogram of v_rest1 and subplot histogram of v_rest2; 
#another subplot if the p-value is >.0.05 two sub plots =ax = plt.subplots(1,1). d) labels x and y axes use ax.ylabel and ax.xlabel

#a means of v_rest1 and 2
mean_v_rest1 = stats.describe(v_rest1)
mean_v_rest2 = stats.describe (v_rest2)

print(mean_v_rest1)
print(mean_v_rest2)

#b t-test
stat = stats.ttest_ind(v_rest1, v_rest2)

#c histogram subplots
fig,ax = plt.subplots(1,2,figsize=(15,5))
ax[0].hist(v_rest1)
ax[1].hist(v_rest2)

#d label axes
ax[0].set_ylabel('Frquency')
ax[1].set_ylabel('Frquency')

ax[0].set_xlabel('Resting Membrane Potential')
ax[1].set_xlabel('Resting Membrane Potential')
```

```python
#Task 5: make a function called vrestHist that executes the code in step 4.
def vrestHist():
  fig, ax = plt.subplots(1,2,figsize=(15,5))
  if stat.pvalue > 0.05:
      ax[0].hist(v_rest1)
      ax[0].set_ylabel('Count')
      ax[0].set_xlabel('Resting Membrance Potential')
      ax[0].set_title('V_Rest 1')

      ax[1].hist(v_rest2)
      ax[1].set_ylabel('Count')
      ax[1].set_xlabel('Resting Membrance Potential')
      ax[1].set_title('V_Rest 2') 
  else:
      ax[0].hist(v_rest1)
      ax[0].set_ylabel('Count')
      ax[0].set_xlabel('Resting Membrance Potential')
      ax[0].set_title('V_Rest 1')

  plt.show()  
```

```python
#Task 6: call the previously created function
vrestHist()
```

```python
#Task 7, make a code cell that loops through the values of v_rest1 and v_rest2 and color codes them depending on how they compare to -60.

v_rest1length = len(v_rest1)

for i in range (v_rest1length):
  if (v_rest1[i] > -60 and v_rest2[i] > -60):
    plt.scatter(v_rest1[i],v_rest2[i],color = "red")
  elif (v_rest1[i] > -60 and v_rest2[i] < -60):
    plt.scatter(v_rest1[i],v_rest2[i], color = "green")
  elif (v_rest1[i] < -60 and v_rest2[i] < -60):
    plt.scatter(v_rest1[i], v_rest2[i], color = "blue")

plt.xlabel ('v_rest1')
plt.ylabel('v_rest2')


```

```python
#Task 8: make a function that makes the previously coded scatterplot

def vrestScatter():
  v_rest1length = len(v_rest1)
  for i in range (v_rest1length):
    if (v_rest1[i] > -60 and v_rest2[i] > -60):
      plt.scatter(v_rest1[i],v_rest2[i],color = "red")
    elif (v_rest1[i] > -60 and v_rest2[i] < -60):
      plt.scatter(v_rest1[i],v_rest2[i], color = "green")
    elif (v_rest1[i] < -60 and v_rest2[i] < -60):
      plt.scatter(v_rest1[i], v_rest2[i], color = "blue")
  plt.xlabel ('v_rest1')
  plt.ylabel('v_rest2')
  plt.show()
```

```python
#Task 9: call the previously established function
vrestScatter()
```
